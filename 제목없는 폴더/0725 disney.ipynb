```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
disney = pd.read_csv('/content/drive/MyDrive/disney.csv')
```

```python
disney['datetime'] = pd.to_datetime(disney['date'])
```

```python
disney.head()
```

```python
disney['year'] = disney['datetime'].dt.year
```

```python
disney['month'] = disney['datetime'].dt.month
```

```python
disney.head()
```

```python
disney2020 = disney[disney['year'] == 2020]
disney2020.head()
```

```python
disney2021 = disney[disney['year'] == 2021]
disney2021.head()
```

```python
disney2022 = disney[disney['year'] == 2022]
disney2022.head()
```

```python
disney1 = disney[disney['month'] == 1]
disney1.head()
```
