```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('/content/drive/MyDrive/netflix_fin.csv')
df
```

```python
df.info()
```

```python
df.at = pd.to_datetime(df['at'])
df.info()
```

```python
df['quarter'] = df['at'].dt.quarter
df.quarter
```

```python
df['year'] = df['at'].dt.strftime('%Y')
df.year
```

```python
df.quarter = df.quarter.astype('str')
```

```python
df.quarter = df.year + '/' + df.quarter + 'Q'
df.quarter
```

```python
df = df.sort_values(by=['at'])
```

```python
plt.subplots(figsize=(30,10))
sns.lineplot(x='quarter', y='score', data=df, ci=None)
plt.show
```
