```python
# ì¿ íŒ¡í”Œë ˆì´
coupang = pd.read_csv('/content/drive/MyDrive/coupang.csv')

reviews = ''

for i in coupang['review'] : 
  reviews = reviews + i + '\n'

coupang_nouns = okt.nouns(reviews)
```

```python
total_list = []
total_list.append(coupang_nouns)
```

```python
from gensim import corpora 
from gensim import models
from wordcloud import WordCloud
import matplotlib.pyplot as plt
```

```python
# dictionary ê°ì²´ ìƒì„±
dictionary = corpora.Dictionary(total_list)
print(dictionary)
```

```python
# Corpus ìƒì„± (ë²¡í„°í™”)
corpus = [dictionary.doc2bow(i) for i in total_list]
```

```python
# lda ëª¨ë¸ ìƒì„±
lda = models.ldamodel.LdaModel(corpus = corpus,
                               id2word = dictionary,
                               num_topics = 20)
```

```python
# ì£¼ì œë§ˆë‹¤ ì¶œí˜„ í™•ë¥ ì´ ë†’ì€ ë‹¨ì–´ìˆœìœ¼ë¡œ ì¶œë ¥
for i in lda.show_topics() :
  print(i)
```
